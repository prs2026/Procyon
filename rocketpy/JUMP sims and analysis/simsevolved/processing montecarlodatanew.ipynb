```python


def quaternion_to_angle_from_vertical(e0, e1, e2, e3):
    """
    Convert quaternion to angle from vertical (zenith angle)
    
    Parameters:
    e0, e1, e2, e3: Quaternion components
    
    Returns:
    angle_degrees: Angle from vertical in degrees
    """
    # Normalize quaternion (just in case)
    norm = np.sqrt(e0**2 + e1**2 + e2**2 + e3**2)
    e0, e1, e2, e3 = e0/norm, e1/norm, e2/norm, e3/norm
    
    # The vertical direction in body frame after rotation
    # Original vertical is [0, 0, 1] (pointing up)
    # Apply quaternion rotation to get current vertical direction
    
    # Quaternion rotation formula for vector [0, 0, 1]
    vx = 2 * (e0*e2 + e1*e3)
    vy = 2 * (e1*e2 - e0*e3)
    vz = e0**2 - e1**2 - e2**2 + e3**2
    
    # Angle from vertical is the angle between [0, 0, 1] and [vx, vy, vz]
    # cos(angle) = dot product with [0, 0, 1] = vz
    angle_rad = np.arccos(np.clip(vz, -1, 1))
    angle_deg = np.degrees(angle_rad)
    
    return angle_deg
```

```python
#sadasdasds

#convert to kml
import math
origin_lat = site_lat  # Replace with your launch site
origin_lon = site_lon

# Convert meters to GPS coordinates
meters_per_degree_lat = 111320
meters_per_degree_lon = 111320 * np.cos(np.radians(origin_lat))

# Landing coordinates
lat_landing = [origin_lat + (y / meters_per_degree_lat) for y in y_data]
lon_landing = [origin_lon + (x / meters_per_degree_lon) for x in x_data]

# Apogee coordinates
lat_apogee = [origin_lat + (y / meters_per_degree_lat) for y in y_apogee]
lon_apogee = [origin_lon + (x / meters_per_degree_lon) for x in x_apogee]

# Calculate statistics for landing
mean_lat_landing = np.mean(lat_landing)
mean_lon_landing = np.mean(lon_landing)
mean_x_landing = np.mean(x_data)
mean_y_landing = np.mean(y_data)
std_x_landing = np.std(x_data)
std_y_landing = np.std(y_data)

# Calculate statistics for apogee
mean_lat_apogee = np.mean(lat_apogee)
mean_lon_apogee = np.mean(lon_apogee)
mean_x_apogee = np.mean(x_apogee)
mean_y_apogee = np.mean(y_apogee)
mean_z_apogee = np.mean(z_apogee)
std_x_apogee = np.std(x_apogee)
std_y_apogee = np.std(y_apogee)
std_z_apogee = np.std(z_apogee)

# Fit each sigma contour independently so its center and orientation can move with the selected samples.
def _xy_points(x_values, y_values):
    x_values = np.asarray(x_values, dtype=float).ravel()
    y_values = np.asarray(y_values, dtype=float).ravel()
    points = np.column_stack((x_values, y_values))
    return points[np.all(np.isfinite(points), axis=1)]


def _sigma_fraction(sigma):
    return min(max(1.0 - math.exp(-0.5 * float(sigma) ** 2), 0.0), 1.0)


def _safe_covariance(points):
    covariance = np.cov(points, rowvar=False, bias=True)
    if covariance.shape != (2, 2) or not np.all(np.isfinite(covariance)):
        covariance = np.eye(2)
    return covariance


def _select_sigma_points(points, sigma):
    if len(points) <= 3:
        return points

    center = np.mean(points, axis=0)
    covariance = _safe_covariance(points)
    jitter = max(float(np.trace(covariance)) * 1e-9, 1e-9)
    inverse_covariance = np.linalg.pinv(covariance + np.eye(2) * jitter)
    deltas = points - center
    distances = np.einsum('ij,jk,ik->i', deltas, inverse_covariance, deltas)

    sample_count = int(np.ceil(_sigma_fraction(sigma) * len(points)))
    sample_count = min(len(points), max(3, sample_count))
    return points[np.argsort(distances)[:sample_count]]


def _ellipse_axes(points):
    center = np.mean(points, axis=0)
    if len(points) < 2:
        return center, np.eye(2), np.array([1.0, 1.0])

    covariance = _safe_covariance(points)
    eigenvalues, eigenvectors = np.linalg.eigh(covariance)
    order = np.argsort(eigenvalues)[::-1]
    eigenvalues = np.maximum(eigenvalues[order], 0.0)
    eigenvectors = eigenvectors[:, order]

    base_radii = np.sqrt(eigenvalues)
    if not np.any(base_radii > 0):
        base_radii = np.array([1.0, 1.0])
    else:
        min_visible_radius = 0.01 * max(float(np.max(base_radii)), 1.0)
        base_radii = np.maximum(base_radii, min_visible_radius)

    return center, eigenvectors, base_radii


def create_sample_sigma_ellipse(x_values, y_values, sigma, altitude=0, num_points=361):
    """Create a closed, independently fitted sigma ellipse in lon/lat coordinates."""
    points = _xy_points(x_values, y_values)
    if len(points) == 0:
        points = np.array([[0.0, 0.0]])

    selected_points = _select_sigma_points(points, sigma)
    center, eigenvectors, base_radii = _ellipse_axes(selected_points)

    projections = (selected_points - center) @ eigenvectors
    normalized_distances = np.sum((projections / np.maximum(base_radii, 1e-9)) ** 2, axis=1)
    enclosing_scale = max(1.0, float(np.sqrt(np.max(normalized_distances))))
    radii = base_radii * enclosing_scale * 1.03

    angles = np.linspace(0, 2 * np.pi, num_points, endpoint=False)
    unit_circle = np.vstack((np.cos(angles), np.sin(angles)))
    offsets = eigenvectors @ (radii[:, None] * unit_circle)

    x_points = center[0] + offsets[0]
    y_points = center[1] + offsets[1]

    # Close the LineString explicitly without relying on a KML Polygon seam.
    x_points = np.append(x_points, x_points[0])
    y_points = np.append(y_points, y_points[0])

    lons = origin_lon + (x_points / meters_per_degree_lon)
    lats = origin_lat + (y_points / meters_per_degree_lat)
    coords = ' '.join([f"{lon},{lat},{altitude}" for lon, lat in zip(lons, lats)])

    major_vector = eigenvectors[:, 0]
    heading_deg = np.degrees(np.arctan2(major_vector[0], major_vector[1])) % 180
    coverage_percent = 100 * len(selected_points) / len(points)

    return coords, center[0], center[1], radii[0], radii[1], heading_deg, len(selected_points), len(points), coverage_percent


# Create KML file
kml_content = f"""<?xml version="1.0" encoding="UTF-8"?>
<kml xmlns="http://www.opengis.net/kml/2.2">
<Document>
    <name>Rocket Flight Data</name>
    <description>Landing points and apogee locations with standard deviation ellipses</description>
    
    <!-- Style for launch site -->
    <Style id="launchStyle">
    <IconStyle>
        <color>ff00ff00</color>
        <scale>1.5</scale>
        <Icon>
        <href>http://maps.google.com/mapfiles/kml/shapes/star.png</href>
        </Icon>
    </IconStyle>
    </Style>
    
    <!-- Style for landing points -->
    <Style id="landingStyle">
      <IconStyle>
        <color>ff0000ff</color>
        <scale>1.0</scale>
        <Icon>
          <href>http://maps.google.com/mapfiles/kml/shapes/placemark_circle.png</href>
        </Icon>
      </IconStyle>
      <LabelStyle>
        <scale>0.3</scale>
        <color>ffffffff</color>
      </LabelStyle>
    </Style>
    
    <!-- Style for apogee points -->
    <Style id="apogeeStyle">
    <IconStyle>
        <color>ffff00ff</color>
        <scale>1.0</scale>
        <Icon>
        <href>http://maps.google.com/mapfiles/kml/shapes/triangle.png</href>
        </Icon>
    </IconStyle>
    <LabelStyle>
        <scale>0.3</scale>
        <color>ffffffff</color>
      </LabelStyle>
    </Style>
    
    <!-- Style for mean landing -->
    <Style id="meanLandingStyle">
    <IconStyle>
        <color>ff00ffff</color>
        <scale>1.3</scale>
        <Icon>
        <href>http://maps.google.com/mapfiles/kml/shapes/target.png</href>
        </Icon>
    </IconStyle>
    </Style>
    
    <!-- Style for mean apogee -->
    <Style id="meanApogeeStyle">
    <IconStyle>
        <color>ffff00aa</color>
        <scale>1.3</scale>
        <Icon>
        <href>http://maps.google.com/mapfiles/kml/shapes/square.png</href>
        </Icon>
    </IconStyle>
    </Style>
    
    <!-- Style for landing std dev ellipse outlines -->
    <Style id="landingSigma1">
    <LineStyle>
        <color>ff0000ff</color>
        <width>3</width>
    </LineStyle>
    <PolyStyle>
        <fill>0</fill>
        <outline>1</outline>
    </PolyStyle>
    </Style>
    
    <Style id="landingSigma2">
    <LineStyle>
        <color>ff0099ff</color>
        <width>3</width>
    </LineStyle>
    <PolyStyle>
        <fill>0</fill>
        <outline>1</outline>
    </PolyStyle>
    </Style>
    
    <Style id="landingSigma3">
    <LineStyle>
        <color>ff00ffff</color>
        <width>3</width>
    </LineStyle>
    <PolyStyle>
        <fill>0</fill>
        <outline>1</outline>
    </PolyStyle>
    </Style>
    
    <!-- Style for apogee std dev ellipse outlines -->
    <Style id="apogeeSigma1">
    <LineStyle>
        <color>ffff00ff</color>
        <width>3</width>
    </LineStyle>
    <PolyStyle>
        <fill>0</fill>
        <outline>1</outline>
    </PolyStyle>
    </Style>
    
    <Style id="apogeeSigma2">
    <LineStyle>
        <color>ffff99ff</color>
        <width>3</width>
    </LineStyle>
    <PolyStyle>
        <fill>0</fill>
        <outline>1</outline>
    </PolyStyle>
    </Style>
    
    <Style id="apogeeSigma3">
    <LineStyle>
        <color>ffffff00</color>
        <width>3</width>
    </LineStyle>
    <PolyStyle>
        <fill>0</fill>
        <outline>1</outline>
    </PolyStyle>
    </Style>
    
    <!-- Launch site -->
    <Placemark>
    <name>Launch Site</name>
    <description>Origin point (0,0)</description>
    <styleUrl>#launchStyle</styleUrl>
    <Point>
        <coordinates>{origin_lon},{origin_lat},0</coordinates>
    </Point>
    </Placemark>
    
    <!-- Mean landing point -->
    <Placemark>
    <name>Mean Landing Point</name>
    <description>Average: {mean_x_landing:.1f}m East, {mean_y_landing:.1f}m North&lt;br/&gt;Std Dev: {std_x_landing:.1f}m x {std_y_landing:.1f}m</description>
    <styleUrl>#meanLandingStyle</styleUrl>
    <Point>
        <coordinates>{mean_lon_landing},{mean_lat_landing},0</coordinates>
    </Point>
    </Placemark>
    
    <!-- Mean apogee point -->
    <Placemark>
    <name>Mean Apogee Point</name>
    <description>Average: {mean_x_apogee:.1f}m East, {mean_y_apogee:.1f}m North, {mean_z_apogee:.1f}m altitude&lt;br/&gt;Std Dev: {std_x_apogee:.1f}m x {std_y_apogee:.1f}m (horizontal), {std_z_apogee:.1f}m (vertical)</description>
    <styleUrl>#meanApogeeStyle</styleUrl>
    <Point>
        <altitudeMode>absolute</altitudeMode>
        <coordinates>{mean_lon_apogee},{mean_lat_apogee},{mean_z_apogee}</coordinates>
    </Point>
    </Placemark>
"""

# Add landing standard deviation ellipses
for sigma, style, name in [(1, 'landingSigma1', 'Landing 1 Std Dev'), 
                            (2, 'landingSigma2', 'Landing 2 Std Dev'), 
                            (3, 'landingSigma3', 'Landing 3 Std Dev')]:
    coords, ellipse_center_x, ellipse_center_y, major_radius, minor_radius, heading_deg, selected_count, total_count, coverage_percent = create_sample_sigma_ellipse(
        x_data, y_data, sigma, altitude=0
    )
    
    kml_content += f"""
    <Placemark>
    <name>{name} Sample-Fitted Ellipse</name>
    <description>{name}: independently fit to closest {selected_count}/{total_count} samples ({coverage_percent:.1f}%); center {ellipse_center_x:.1f}m E, {ellipse_center_y:.1f}m N; major radius {major_radius:.1f}m, minor radius {minor_radius:.1f}m, heading {heading_deg:.1f} deg clockwise from north</description>
    <styleUrl>#{style}</styleUrl>
    <LineString>
        <tessellate>1</tessellate>
        <altitudeMode>clampToGround</altitudeMode>
        <coordinates>
            {coords}
        </coordinates>
    </LineString>
    </Placemark>
"""

# Add apogee standard deviation ellipses
for sigma, style, name in [(1, 'apogeeSigma1', 'Apogee 1 Std Dev'), 
                            (2, 'apogeeSigma2', 'Apogee 2 Std Dev'), 
                            (3, 'apogeeSigma3', 'Apogee 3 Std Dev')]:
    coords, ellipse_center_x, ellipse_center_y, major_radius, minor_radius, heading_deg, selected_count, total_count, coverage_percent = create_sample_sigma_ellipse(
        x_apogee, y_apogee, sigma, altitude=mean_z_apogee
    )
    
    kml_content += f"""
    <Placemark>
    <name>{name} Sample-Fitted Ellipse</name>
    <description>{name}: independently fit to closest {selected_count}/{total_count} samples ({coverage_percent:.1f}%); center {ellipse_center_x:.1f}m E, {ellipse_center_y:.1f}m N; major radius {major_radius:.1f}m, minor radius {minor_radius:.1f}m, heading {heading_deg:.1f} deg clockwise from north at ~{mean_z_apogee:.1f}m altitude</description>
    <styleUrl>#{style}</styleUrl>
    <LineString>
        <tessellate>1</tessellate>
        <altitudeMode>absolute</altitudeMode>
        <coordinates>
            {coords}
        </coordinates>
    </LineString>
    </Placemark>
"""

# Add all landing points
for i, (lon, lat, x, y) in enumerate(zip(lon_landing, lat_landing, x_data, y_data), 1):
    kml_content += f"""
    <Placemark>
    <name>Landing {i}</name>
    <description>Offset: {x:.1f}m East, {y:.1f}m North</description>
    <styleUrl>#landingStyle</styleUrl>
    <Point>
        <coordinates>{lon},{lat},0</coordinates>
    </Point>
    </Placemark>
"""

# Add all apogee points
for i, (lon, lat, x, y, z) in enumerate(zip(lon_apogee, lat_apogee, x_apogee, y_apogee, z_apogee), 1):
    kml_content += f"""
    <Placemark>
    <name>Apogee {i}</name>
    <description>Position: {x:.1f}m East, {y:.1f}m North&lt;br/&gt;Altitude: {z:.1f}m</description>
    <styleUrl>#apogeeStyle</styleUrl>
    <Point>
        <altitudeMode>absolute</altitudeMode>
        <coordinates>{lon},{lat},{z}</coordinates>
    </Point>
    </Placemark>
"""

kml_content += """
</Document>
</kml>
"""

# Save to file with UTF-8 encoding
filename = "rocket_flight_data.kml"
with open(filename, 'w', encoding='utf-8') as f:
    f.write(kml_content)

print(f"KML file saved as '{filename}'")
print(f"\nLaunch site: {origin_lat:.6f}, {origin_lon:.6f}")
print(f"\nLanding Statistics:")
print(f"  Mean: {mean_lat_landing:.6f}, {mean_lon_landing:.6f} ({mean_x_landing:.1f}m E, {mean_y_landing:.1f}m N)")
print(f"  Std Dev: {std_x_landing:.1f}m (E-W) x {std_y_landing:.1f}m (N-S)")
print(f"\nApogee Statistics:")
print(f"  Mean: {mean_lat_apogee:.6f}, {mean_lon_apogee:.6f} ({mean_x_apogee:.1f}m E, {mean_y_apogee:.1f}m N)")
print(f"  Mean Altitude: {mean_z_apogee:.1f}m +/- {std_z_apogee:.1f}m")
print(f"  Horizontal Std Dev: {std_x_apogee:.1f}m (E-W) x {std_y_apogee:.1f}m (N-S)")
print(f"\nTotal flights: {len(lat_landing)}")
```

```python
import numpy as np
import math


def create_flight_paths_kml(flight_results, origin_lat, origin_lon, filename='flight_paths.kml'):
    """
    Create a KML file with all flight trajectories from RocketPy simulations.
    
    Parameters:
    -----------
    flight_results : list
        List of dictionaries containing flight data. Each dict should have:
        - 'x_path': list/array of x positions over time (meters East)
        - 'y_path': list/array of y positions over time (meters North)
        - 'z_path': list/array of altitudes over time (meters)
        - Optional: 'flight_number' or other identifying info
    origin_lat : float
        Launch site latitude
    origin_lon : float
        Launch site longitude
    filename : str
        Output KML filename
    
    Returns:
    --------
    str : Path to created KML file
    """
    
    # Conversion constants
    meters_per_degree_lat = 111320
    meters_per_degree_lon = 111320 * np.cos(np.radians(origin_lat))

    def offset_to_lon_lat(x, y):
        return origin_lon + (x / meters_per_degree_lon), origin_lat + (y / meters_per_degree_lat)

    def finite_xy_points(points):
        points = np.asarray(points, dtype=float)
        if points.ndim == 1:
            points = points.reshape(-1, 2)
        else:
            points = points[:, :2]
        return points[np.all(np.isfinite(points), axis=1)]


    def sigma_fraction(sigma):
        return min(max(1.0 - math.exp(-0.5 * float(sigma) ** 2), 0.0), 1.0)


    def safe_covariance(points):
        covariance = np.cov(points, rowvar=False, bias=True)
        if covariance.shape != (2, 2) or not np.all(np.isfinite(covariance)):
            covariance = np.eye(2)
        return covariance


    def select_sigma_points(points, sigma):
        if len(points) <= 3:
            return points

        center = np.mean(points, axis=0)
        covariance = safe_covariance(points)
        jitter = max(float(np.trace(covariance)) * 1e-9, 1e-9)
        inverse_covariance = np.linalg.pinv(covariance + np.eye(2) * jitter)
        deltas = points - center
        distances = np.einsum('ij,jk,ik->i', deltas, inverse_covariance, deltas)

        sample_count = int(np.ceil(sigma_fraction(sigma) * len(points)))
        sample_count = min(len(points), max(3, sample_count))
        return points[np.argsort(distances)[:sample_count]]


    def ellipse_axes(points):
        center = np.mean(points, axis=0)
        if len(points) < 2:
            return center, np.eye(2), np.array([1.0, 1.0])

        covariance = safe_covariance(points)
        eigenvalues, eigenvectors = np.linalg.eigh(covariance)
        order = np.argsort(eigenvalues)[::-1]
        eigenvalues = np.maximum(eigenvalues[order], 0.0)
        eigenvectors = eigenvectors[:, order]

        base_radii = np.sqrt(eigenvalues)
        if not np.any(base_radii > 0):
            base_radii = np.array([1.0, 1.0])
        else:
            min_visible_radius = 0.01 * max(float(np.max(base_radii)), 1.0)
            base_radii = np.maximum(base_radii, min_visible_radius)

        return center, eigenvectors, base_radii


    def create_sample_sigma_ellipse(points, sigma, altitude=0, num_points=361):
        """Create a closed, independently fitted sigma ellipse in lon/lat coordinates."""
        points = finite_xy_points(points)
        if len(points) == 0:
            points = np.array([[0.0, 0.0]])

        selected_points = select_sigma_points(points, sigma)
        center, eigenvectors, base_radii = ellipse_axes(selected_points)

        projections = (selected_points - center) @ eigenvectors
        normalized_distances = np.sum((projections / np.maximum(base_radii, 1e-9)) ** 2, axis=1)
        enclosing_scale = max(1.0, float(np.sqrt(np.max(normalized_distances))))
        radii = base_radii * enclosing_scale * 1.03

        angles = np.linspace(0, 2 * np.pi, num_points, endpoint=False)
        unit_circle = np.vstack((np.cos(angles), np.sin(angles)))
        offsets = eigenvectors @ (radii[:, None] * unit_circle)

        x_points = center[0] + offsets[0]
        y_points = center[1] + offsets[1]

        # Close the LineString explicitly without relying on a KML Polygon seam.
        x_points = np.append(x_points, x_points[0])
        y_points = np.append(y_points, y_points[0])

        coords = []
        for x, y in zip(x_points, y_points):
            lon, lat = offset_to_lon_lat(x, y)
            coords.append(f"{lon},{lat},{altitude}")

        major_vector = eigenvectors[:, 0]
        heading_deg = np.degrees(np.arctan2(major_vector[0], major_vector[1])) % 180
        coverage_percent = 100 * len(selected_points) / len(points)

        return ' '.join(coords), center[0], center[1], radii[0], radii[1], heading_deg, len(selected_points), len(points), coverage_percent


    landing_points = []
    for flight in flight_results:
        if len(flight['x_path']) == 0 or len(flight['y_path']) == 0:
            continue
        landing_x = float(flight['x_path'][-1])
        landing_y = float(flight['y_path'][-1])
        if np.isfinite(landing_x) and np.isfinite(landing_y):
            landing_points.append((landing_x, landing_y))

    if landing_points:
        landing_points_array = np.asarray(landing_points, dtype=float)
        mean_landing_x = np.mean(landing_points_array[:, 0])
        mean_landing_y = np.mean(landing_points_array[:, 1])
        mean_landing_lon, mean_landing_lat = offset_to_lon_lat(mean_landing_x, mean_landing_y)
    
    # Start KML content
    kml_content = f"""<?xml version="1.0" encoding="UTF-8"?>
<kml xmlns="http://www.opengis.net/kml/2.2">
  <Document>
    <name>Rocket Flight Trajectories</name>
    <description>All flight paths from simulations</description>
    
    <!-- Style for launch site -->
    <Style id="launchStyle">
      <IconStyle>
        <color>ff00ff00</color>
        <scale>2.0</scale>
        <Icon>
          <href>http://maps.google.com/mapfiles/kml/shapes/star.png</href>
        </Icon>
      </IconStyle>
      <LabelStyle>
        <scale>1.5</scale>
        <color>ffffffff</color>
      </LabelStyle>
    </Style>
    
    <!-- Style for flight path -->
    <Style id="flightPath">
      <LineStyle>
        <color>ff0000ff</color>
        <width>3</width>
      </LineStyle>
    </Style>
    
    <!-- Style for apogee marker -->
    <Style id="apogeeStyle">
      <IconStyle>
        <color>ffff00ff</color>
        <scale>0.8</scale>
        <Icon>
          <href>http://maps.google.com/mapfiles/kml/shapes/triangle.png</href>
        </Icon>
      </IconStyle>
      <LabelStyle>
        <scale>0.3</scale>
        <color>ffffffff</color>
      </LabelStyle>
    </Style>
    
    <!-- Style for landing marker -->
    <Style id="landingStyle">
      <IconStyle>
        <color>ff0000ff</color>
        <scale>0.3</scale>
        <Icon>
          <href>http://maps.google.com/mapfiles/kml/shapes/placemark_circle.png</href>
        </Icon>
      </IconStyle>
      <LabelStyle>
        <scale>0.7</scale>
        <color>ffffffff</color>
      </LabelStyle>
    </Style>

    <!-- Style for mean landing marker -->
    <Style id="meanLandingStyle">
      <IconStyle>
        <color>ff00ffff</color>
        <scale>1.0</scale>
        <Icon>
          <href>http://maps.google.com/mapfiles/kml/shapes/target.png</href>
        </Icon>
      </IconStyle>
      <LabelStyle>
        <scale>0.8</scale>
        <color>ffffffff</color>
      </LabelStyle>
    </Style>

    <!-- Styles for landing sigma ellipse outlines -->
    <Style id="landingSigma1">
      <LineStyle>
        <color>ff0000ff</color>
        <width>3</width>
      </LineStyle>
      <PolyStyle>
        <fill>0</fill>
        <outline>1</outline>
      </PolyStyle>
    </Style>

    <Style id="landingSigma2">
      <LineStyle>
        <color>ff0099ff</color>
        <width>3</width>
      </LineStyle>
      <PolyStyle>
        <fill>0</fill>
        <outline>1</outline>
      </PolyStyle>
    </Style>

    <Style id="landingSigma3">
      <LineStyle>
        <color>ff00ffff</color>
        <width>3</width>
      </LineStyle>
      <PolyStyle>
        <fill>0</fill>
        <outline>1</outline>
      </PolyStyle>
    </Style>
    
    <!-- Launch site marker -->
    <Placemark>
      <name>Launch Site</name>
      <description>Common launch point for all flights</description>
      <styleUrl>#launchStyle</styleUrl>
      <Point>
        <coordinates>{origin_lon},{origin_lat},0</coordinates>
      </Point>
    </Placemark>
"""

    if landing_points:
        kml_content += f"""
    <!-- Mean landing point -->
    <Placemark>
      <name>Mean Landing Point</name>
      <description>Average landing position: {mean_landing_x:.1f}m E, {mean_landing_y:.1f}m N</description>
      <styleUrl>#meanLandingStyle</styleUrl>
      <Point>
        <coordinates>{mean_landing_lon},{mean_landing_lat},0</coordinates>
      </Point>
    </Placemark>
"""

        for sigma, style, name in [(1, 'landingSigma1', 'Landing 1 Sigma'),
                                   (2, 'landingSigma2', 'Landing 2 Sigma'),
                                   (3, 'landingSigma3', 'Landing 3 Sigma')]:
            coords, ellipse_center_x, ellipse_center_y, major_radius, minor_radius, heading_deg, selected_count, total_count, coverage_percent = create_sample_sigma_ellipse(
                landing_points_array, sigma, altitude=0
            )
            kml_content += f"""
    <!-- {name} sample-fitted ellipse -->
    <Placemark>
      <name>{name} Sample-Fitted Ellipse</name>
      <description>{name}: independently fit to closest {selected_count}/{total_count} samples ({coverage_percent:.1f}%); center {ellipse_center_x:.1f}m E, {ellipse_center_y:.1f}m N; major radius {major_radius:.1f}m, minor radius {minor_radius:.1f}m, heading {heading_deg:.1f} deg clockwise from north</description>
      <styleUrl>#{style}</styleUrl>
      <LineString>
        <tessellate>1</tessellate>
        <altitudeMode>clampToGround</altitudeMode>
        <coordinates>
          {coords}
        </coordinates>
      </LineString>
    </Placemark>
"""
    
    # Color palette for different flights (AABBGGRR format)
    colors = [
        'ff0000ff',  # Red
        'ff00ff00',  # Green
        'ffff0000',  # Blue
        'ff00ffff',  # Yellow
        'ffff00ff',  # Magenta
        'ffffff00',  # Cyan
        'ff0080ff',  # Orange
        'ff800080',  # Purple
        'ff008080',  # Olive
        'ff80ff80',  # Light green
    ]
    
    # Process each flight
    for i, flight in enumerate(flight_results):
        flight_num = i + 1
        color = colors[i % len(colors)]
        
        # Get flight path data
        x_path = flight['x_path']
        y_path = flight['y_path']
        z_path = flight['z_path']
        
        # Convert to lat/lon
        lat_path = [origin_lat + (y / meters_per_degree_lat) for y in y_path]
        lon_path = [origin_lon + (x / meters_per_degree_lon) for x in x_path]
        
        # Find apogee (maximum altitude)
        apogee_idx = np.argmax(z_path)
        apogee_alt = z_path[apogee_idx]
        apogee_lat = lat_path[apogee_idx]
        apogee_lon = lon_path[apogee_idx]
        apogee_x = x_path[apogee_idx]
        apogee_y = y_path[apogee_idx]
        
        # Landing point
        landing_lat = lat_path[-1]
        landing_lon = lon_path[-1]
        landing_x = x_path[-1]
        landing_y = y_path[-1]
        
        # Create style for this flight path
        kml_content += f"""
    <!-- Style for flight {flight_num} -->
    <Style id="flight{flight_num}Path">
      <LineStyle>
        <color>{color}</color>
        <width>2</width>
      </LineStyle>
    </Style>
"""
        
        # Create coordinates string for path
        coords = ' '.join([f"{lon},{lat},{alt}" 
                          for lon, lat, alt in zip(lon_path, lat_path, z_path)])
        
        # Add flight path
        kml_content += f"""
    <!-- Flight {flight_num} trajectory -->
    <Placemark>
      <name>Flight {flight_num} Path</name>
      <description>
        Apogee: {apogee_alt:.1f}m at ({apogee_x:.1f}m E, {apogee_y:.1f}m N)&lt;br/&gt;
        Landing: ({landing_x:.1f}m E, {landing_y:.1f}m N)
      </description>
      <styleUrl>#flight{flight_num}Path</styleUrl>
      <LineString>
        <extrude>0</extrude>
        <tessellate>1</tessellate>
        <altitudeMode>absolute</altitudeMode>
        <coordinates>
          {coords}
        </coordinates>
      </LineString>
    </Placemark>
"""
        
        # Add apogee marker
        kml_content += f"""
    <!-- Flight {flight_num} apogee -->
    <Placemark>
      <name>F{flight_num} Apogee</name>
      <description>
        Altitude: {apogee_alt:.1f}m&lt;br/&gt;
        Position: {apogee_x:.1f}m E, {apogee_y:.1f}m N
      </description>
      <styleUrl>#apogeeStyle</styleUrl>
      <Point>
        <altitudeMode>absolute</altitudeMode>
        <coordinates>{apogee_lon},{apogee_lat},{apogee_alt}</coordinates>
      </Point>
    </Placemark>
"""
        
        # Add landing marker
        kml_content += f"""
    <!-- Flight {flight_num} landing -->
    <Placemark>
      <name>F{flight_num} Landing</name>
      <description>
        Position: {landing_x:.1f}m E, {landing_y:.1f}m N
      </description>
      <styleUrl>#landingStyle</styleUrl>
      <Point>
        <coordinates>{landing_lon},{landing_lat},0</coordinates>
      </Point>
    </Placemark>
"""
    
    # Close KML document
    kml_content += """
  </Document>
</kml>
"""
    
    # Save to file
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(kml_content)
    
    print(f"KML file saved as '{filename}'")
    print(f"  Total flights: {len(flight_results)}")
    print(f"  Launch site: {origin_lat:.6f}, {origin_lon:.6f}")
    
    return filename


# ============ EXAMPLE USAGE ============

# Example 1: If you have the full Flight objects from RocketPy
def convert_solutions_to_flight_paths(solutions, min_landing_time=10):
    """
    Convert simulation solutions to flight path format for KML generation.
    
    Parameters:
    -----------
    solutions : list
        List of solution arrays. Each solution is a 2D array where:
        solution[time_step] = [t, x, y, z, vx, vy, vz, e0, e1, e2, e3, omega_x, omega_y, omega_z]
    min_landing_time : float or None
        Skip solutions whose final timestamp is at or below this value. Defaults to 10 seconds to match the landing-point KML export.
        
    Returns:
    --------
    list : List of dictionaries with format:
        {
            'x_path': list of x positions (m),
            'y_path': list of y positions (m),
            'z_path': list of z positions (m),
            'flight_number': int
        }
    """
    flight_results = []
    
    for i, solution in enumerate(solutions):
        # Convert to numpy array if not already
        solution = np.array(solution)
        if len(solution) == 0:
            continue
        if min_landing_time is not None and solution[-1, 0] <= min_landing_time:
            continue
        
        # Extract position columns
        x_path = solution[:, 1].tolist()  # Column 1: X position
        y_path = solution[:, 2].tolist()  # Column 2: Y position
        z_path = solution[:, 3].tolist()  # Column 3: Z position
        
        flight_results.append({
            'x_path': x_path,
            'y_path': y_path,
            'z_path': z_path,
            'flight_number': i + 1
        })
    
    return flight_results


# ============ EXAMPLE USAGE ============

# Example: You have a list of solutions from your simulations
# solutions = [solution1, solution2, solution3, ...]
# where each solution is a 2D array with columns [t, x, y, z, vx, vy, vz, ...]

# Example 2: If you already have arrays of positions
# Simulated example data
# flight_results = []
# for i in range(5):
#     # Create sample trajectory
#     t = np.linspace(0, 60, 100)  # 60 seconds, 100 points
    
#     # Parabolic trajectory with some randomness
#     drift_x = np.random.uniform(-2, 2)
#     drift_y = np.random.uniform(-2, 2)
    
#     x_path = t * drift_x + np.random.normal(0, 5, len(t))
#     y_path = t * drift_y + np.random.normal(0, 5, len(t))
#     z_path = -9.81/2 * (t - 30)**2 + 1000 + np.random.normal(0, 10)
#     z_path = np.maximum(z_path, 0)  # Don't go below ground
    
#     flight_results.append({
#         'x_path': x_path.tolist(),
#         'y_path': y_path.tolist(),
#         'z_path': z_path.tolist()
#     })

plotables = convert_solutions_to_flight_paths(results)

# Create the KML
origin_lat = site_lat
origin_lon = site_lon

create_flight_paths_kml(plotables, origin_lat, origin_lon, 'flight_trajectories.kml')
```
